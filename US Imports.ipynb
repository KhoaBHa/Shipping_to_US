```python
import pandas as pd
df = pd.read_excel(r"C:\Users\habao\OneDrive\Documents\Data Analysis\Porfolio\US Import\US Import_Tableau Viz.xlsx")
df
```

```python
#Total sales from each department
df.groupby(['Department Name']).sum(numeric_only=True).filter(items = ['total_sale']).sort_values(by=['total_sale'], ascending = False)
```

```python
#Total sales from each product
df.groupby(['Product Name']).sum(numeric_only=True).filter(items = ['total_sale']).sort_values(by=['total_sale'], ascending = False)
```

```python
#Highest discount rate from each department
df.groupby(['Department Name']).max(numeric_only=True).filter(items = ['discount_rate']).sort_values(by=['discount_rate'], ascending = False)
```

```python
#Numbers of late delivery by departments
df[df['Late_delivery_risk'] == 1].groupby(['Department Name']).count().filter(items = ['Late_delivery_risk'])
```

```python
#Total sales created from each state
df.groupby('Customer State').sum(numeric_only=True).filter(items = ['Customer State','total_sale']).sort_values(by='total_sale', ascending=False)
```

```python
#Top 10 states with the highest total import
df.groupby('Customer State').sum(numeric_only=True).filter(items = ['Customer State','total_sale']).sort_values(by='total_sale', ascending=False).head(10)
```

```python
#Preferred shipping mode of each payment type
df.groupby(['payment_type','Shipping Mode']).count().filter(items = ['payment_type','Shipping Mode','shipping_date']).sort_values(by = ['payment_type','shipping_date','Shipping Mode',], ascending=[True,False,True])
```
